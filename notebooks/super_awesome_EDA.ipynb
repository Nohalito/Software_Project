```python
# Base import
import pandas as pd

# Visualization
import seaborn as sns
import matplotlib.pyplot as plt

#Repo management
import sys
sys.path.append('..')

import config as c

# Code aesthetic
#import tqdm
import warnings

warnings.filterwarnings('ignore')
```

```python
path = '../' + c.PROCESSED_DATA_PATH + c.PER_GAME_DF_FILENAME

df = pd.read_csv(path,
                index_col = 0)

print(f"The downloaded dataset have {df.shape[0]} rows & {df.shape[1]} columns.")
df.head()
```

```python
df_2020 = df.loc[df['Year'] == 2020]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_2020['PTS'], bins=30, kde=True)
plt.title('Distribution of Points Per Game in 2020')
plt.xlabel('Points')
plt.ylabel('Number of Players')
plt.show()
```

```python
sns.boxplot(x='Pos', y='PTS', data=df_2020)
plt.title('Points Per Game by Position in 2020')
plt.show()
```

```python
sns.scatterplot(
    x='PTS', y='FG%', hue='Pos', size='MP',
    data=df_2020, alpha=0.7
)
plt.title('Scoring vs Field Goal Percentage in 2020')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(
    df_2020[['PTS','AST','MP','FG%','3P%','FT%']].corr(),
    annot=True, cmap='coolwarm'
)
plt.title('Correlation Between Player Stats')
plt.show()
```

```python
sns.scatterplot(x='Age', y='PTS', data=df_2020)
plt.title('Age vs Points Per Game in 2020')
plt.show()
```

```python
sns.scatterplot(x='MP', y='PTS', data=df_2020)
plt.title('Minutes Played vs Points in 2020')
plt.show()
```

```python
sns.pairplot(
    df_2020[['PTS','AST','MP','FG%']],
    corner=True
)
plt.show()
```

```python
top_scorers = df_2020.sort_values('PTS', ascending=False).head(10)

sns.barplot(x='PTS', y='Player', data=top_scorers)
plt.title('Top Scorers in 2020')
plt.show()
```
